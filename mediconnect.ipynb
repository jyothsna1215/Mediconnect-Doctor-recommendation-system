```python
import pandas as pd
df["disease_list"] = df["disease_name"].apply(lambda x: x if isinstance(x, list) else x.split(","))
def rec(disease):
    try:
        disease = disease.strip().lower()
        matching_doctors = df[df["disease_list"].apply(lambda x: disease in [d.strip().lower() for d in x])]

        if matching_doctors.empty:
            return "No doctors found for the given disease."

        recommendations = matching_doctors.apply(
            lambda row: f"Doctor: {row['doctor_name']} | Hospital: {row['hospital_name']} | Location: {row['location']} | Contact: {row['contact_number']}",
            axis=1
        ).tolist()

        return "\n".join(recommendations)
    except Exception as e:
        return str(e)
user_input = input("Enter the disease name: ")
print("\nRecommendations:\n", rec(user_input))
```
